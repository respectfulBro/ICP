```python
def is_symmetric(A):
    for i in range(len(A)):
        for j in range(len(A[i])):
            if i!=j and A[i][j]!=A[j][i]:
                return False
    return True
```

1. Check if a Matrix is Square
Write a function is_square(A) that checks if a matrix is a square matrix. A matrix is square if the number of rows equals the number of columns.

```python
def is_square(A):
    rows=len(A)
    for row in rows:
        if len(row)!=rows:
            return False
    return True
```
